# Let's analyse the RGB Format 

```python
img1 =cv2.imread('images/elephant.jpg')
cv2.imshow('image',img1)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img1.shape
```

Splitting the color component of the image in BGR


```python
B,G,R =cv2.split(img1)
```

```python
print("shape os red channel", R.shape)
print("shape os green channel", G.shape)
print("shape os blue channel", B.shape)
```

```python
cv2.imshow('image in BGR',img1)
cv2.imshow('Vlue channel',B)
cv2.imshow('Green channel',G)
cv2.imshow('Red channel',R)
cv2.waitKey()
cv2.destroyAllWindows()
```

Merging the color component into RGB form and analyse that

```python
merger_RGB =cv2.merge([R,B,G])
merger_BRG =cv2.merge([B,R,G])
cv2.imshow("Merge RGB", merger_RGB)
cv2.imshow("Merge BRG",merger_BRG)
cv2.imshow("Simple BGR", img1)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#let's amplify the blue color
merger_amplify =cv2.merge([B+100,G,R])
cv2.imshow("Merge BGR amplify blue",merger_amplify)
cv2.imshow("Simple BGR", img1)
cv2.waitKey()
cv2.destroyAllWindows()
```

Hence in the above way we can analyse different font of color form

# Histogram way to analyse color component

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image =cv2.imread("images/elephant.jpg")
image.shape
```

```python
histogram =cv2.calcHist([image], [0], None, [256], [0,256])

# ravel function will help in flatten i.e convet 2d array to 1d
plt.hist(image.ravel(), 256, [0,256])
plt.plot(histogram, color='b')
plt.show()
```

seprate color component representation of BGR

```python
color =('b', 'g', 'r')

for i,col in enumerate(color):
    histogram1 =cv2.calcHist([image], [i], None, [256], [0,256])
    plt.plot(histogram1, color=col)
    
plt.show()
```
